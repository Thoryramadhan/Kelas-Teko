### Soal 3

```python
def arctan_taylor(x, N):
    total = 0.0
    for n in range(N + 1):
        total += ((-1) ** n) * (x ** (4 * n + 2)) / (2 * n + 1)
    return total

# Uji fungsi arctan_taylor untuk x=1, N=5000
val = arctan_taylor(1, N)
print("arctan(1^2)     :", val)
```

### Soal 4

```python
def sqrt_integral(u, p, N):

    dx = (p - u) / N
    total = 0.0
    for i in range(N):
        x_mid = u + (i + 0.5) * dx
        total += math.sqrt(1 + x_mid**4)
    return total * dx

if __name__ == "__main__":
    u = 0
    p = 1
    N = 1000
    I_approx = sqrt_integral(u, p, N)
    print(f"Pendekatan integral sqrt dari {u} hingga {p} =", I_approx)
```

### Soal 5

```python
# Program sederhana untuk menghitung invers matriks 2x2

# Minta input dari pengguna
a1 = float(input("Masukkan nilai a1: "))
a2 = float(input("Masukkan nilai a2: "))
a3 = float(input("Masukkan nilai a3: "))
a4 = float(input("Masukkan nilai a4: "))

# Bentuk matriks 2x2
A = [[a1, a2],
     [a3, a4]]

# Tampilkan matriks A
print("\nMatriks A:")
for baris in A:
    print(baris)

# Hitung determinan
det = a1 * a4 - a2 * a3

# Jika determinan = 0, matriks tidak memiliki invers
if det == 0:
    print("\nMatriks tidak memiliki invers (determinannya nol).")
else:
    # Rumus invers 2x2
    A_inv = [
        [ a4/det, -a2/det],
        [-a3/det,  a1/det]
    ]
    print("\nInvers matriks A:")
    for baris in A_inv:
        print(baris)
```

### Soal 6

```python
def root_sequence(N, b, x0):

    # Inisialisasi list dengan suku awal
    x = [x0]
    
    # Bangun barisan hingga suku ke-(N-1)
    for n in range(1, N):
        x_next = b + 1.0 / x[n-1]
        x.append(x_next)
    
    return x

# Contoh penggunaan
if __name__ == "__main__":
    seq = root_sequence(N=5, b=1, x0=1)  # misalnya 5 suku, b=1, x0=1
    print("Hasil root_sequence:", seq)
```

### Soal 7

```python
def fib_less_than(N):
    fibs = []
    a, b = 0, 1
    while a < N:
        fibs.append(a)
        a, b = b, a + b
    return fibs

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def fibonacci_primes(N):
    """
    Mengembalikan list bilangan Fibonacci < N
    yang juga merupakan bilangan prima.
    """
    fibs = fib_less_than(N)
    return [f for f in fibs if is_prime(f)]

# Contoh penggunaan
if __name__ == "__main__":
    print("Fibonacci < 30:", fib_less_than(30))
    print("Fibonacci prima < 30:", fibonacci_primes(30))
```

### Soal 8

```python
def dotprod(M, i, j):
    n = len(M)          # jumlah baris
    # atau len(M[0])     # jumlah kolom, diasumsikan sama
    result = 0
    for k in range(n):
        result += M[i][k] * M[k][j]
    return result

# Contoh penggunaan
if __name__ == "__main__":
    M = [
        [1, 2],
        [3, 4]
    ]
    print("M =", M)
    print("dotprod(M, 0, 1) =", dotprod(M, 0, 1))  # Baris 0 . Kolom 1
```

### Soal 9

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_three_functions():
    # Data untuk f(x) = e^x
    x1 = np.linspace(-1, 1, 100)
    f = np.exp(x1)
    
    # Data untuk g(x) = 1 + sin(2x)
    x2 = np.linspace(0, 4*np.pi, 200)
    g = 1 + np.sin(2*x2)
    
    # Data untuk h(x) = 2 - 2 sin^2(3πx/2)
    x3 = np.linspace(-3, 3, 300)
    h = 2 - 2 * (np.sin((3*np.pi*x3)/2))**2
    
    # Membuat figure dan 3 subplot (1 baris, 3 kolom)
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    
    # Subplot 1: f(x)
    axes[0].plot(x1, f, 'b')
    axes[0].set_title("f(x) = e^x")
    axes[0].set_xlabel("x")
    axes[0].set_ylabel("f(x)")
    
    # Subplot 2: g(x)
    axes[1].plot(x2, g, 'r')
    axes[1].set_title("g(x) = 1 + sin(2x)")
    axes[1].set_xlabel("x")
    axes[1].set_ylabel("g(x)")
    
    # Subplot 3: h(x)
    axes[2].plot(x3, h, 'g')
    axes[2].set_title("h(x) = 2 - 2 sin^2(3πx/2)")
    axes[2].set_xlabel("x")
    axes[2].set_ylabel("h(x)")
    
    plt.tight_layout()
    plt.show()

# Contoh pemanggilan fungsi
if __name__ == "__main__":
    plot_three_functions()
```

### Soal 10

```python
import numpy as np
import matplotlib.pyplot as plt

def parametric_plots(a, k):
    """
    Membuat dua plot parametrik dalam 2 subplot:
      1) x(t) = 2k cos(t), y(t) = a cos(kt)
      2) x(t) = 2k sin(t), y(t) = a sin(kt)
    dengan t dari 0 s/d 2π.
    """
    t = np.linspace(0, 2*np.pi, 300)
    
    # Parametric 1
    x1 = 2*k * np.cos(t)
    y1 = a * np.cos(k * t)
    
    # Parametric 2
    x2 = 2*k * np.sin(t)
    y2 = a * np.sin(k * t)
    
    # Buat figure dengan 2 subplot
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    
    # Subplot 1
    axes[0].plot(x1, y1, 'b')
    axes[0].set_title(f"x=2k cos(t), y=a cos(kt)\n(a={a}, k={k})")
    axes[0].axis("equal")  # Supaya skala x dan y sama
    
    # Subplot 2
    axes[1].plot(x2, y2, 'r')
    axes[1].set_title(f"x=2k sin(t), y=a sin(kt)\n(a={a}, k={k})")
    axes[1].axis("equal")  # Supaya skala x dan y sama
    
    plt.tight_layout()
    plt.show()

# Contoh pemanggilan fungsi
if __name__ == "__main__":
    parametric_plots(a=2, k=3)
```
